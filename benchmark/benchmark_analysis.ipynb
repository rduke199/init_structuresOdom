```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cwd = os.getcwd()
datapath = os.path.join(cwd,'master.json')
benchmark_data = pd.read_json(datapath)
data = pd.json_normalize(benchmark_data['molecules'])

ip_dict = {'benzene' : 9.24, 
           'anisole' : 8.2,
           'cprop_cation' : 6.6,
           'anthracene' : 7.44,
           'aniline' : 7.72, 
           'biphenyl' : 8.16
          }

def FindIP(molecule_name): 
    ip = ip_dict[molecule_name]
    return ip

data['ionization_potential'] = data['molecule_name'].apply(FindIP)
data['cenergy_nenergy'] = data['cation_energy'] - data['energy']
data['ip_error'] = data['ionization_potential'] - data['cenergy_nenergy']
data['percent_error'] = abs(data['ip_error'] / data['ionization_potential'])

# data.to_csv('master.csv')
data
```

```python
df_ncprop = data
df_ncprop.drop(df_ncprop[df_ncprop['molecule_name'] == 'cprop_cation'].index, inplace = True) 
df_ncprop
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
funcitonals = ['LC-wHPBE','wB97XD']
basis_sets = ['cc-pVDZ', '6-31+G**', 'TZVP', 'aug-cc-pVTZ']

grouped_data = df_ncprop.groupby(by='functional')
for f_group, f_dat in grouped_data:
    bs_group = grouped_data.get_group(f_group).groupby(by='basis_set')
    
    df = bs_group['runtime'].mean()
    df.plot.bar()
    plt.title(f_group)
    plt.ylim(top = 1,bottom = 0)
    
    f_dat.boxplot(column='percent_error', by='basis_set')
    plt.title(f_group)
    plt.suptitle("")
    plt.ylim(top = 0.10 ,bottom = 0)
    plt.show()
    
    for bs_group, bs_dat in bs_group:         
        rms_error = rmse(bs_dat.cenergy_nenergy,bs_dat.ionization_potential)
        print("{}, {} \t{}".format(f_group,bs_group,rms_error))
#         print(bs_dat[['molecule_name','cenergy_nenergy','ionization_potential']])
```
